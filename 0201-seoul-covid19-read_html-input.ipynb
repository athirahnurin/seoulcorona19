```python
# read_html 은 사이트 내의 html 태그를 불러옵니다.

table = pd.read_html(url)
```

```python
# table 변수 안에 있는 table 태그의 값을 하나씩 읽어보며 확진자 목록이 있는 테이블을 가져옵니다.
df=table[3]
df.shape
```

```python
# 미리보기 합니다.
df.head()
```

```python
# 마지막 확진일을 파일명에 써주기 위해 . 을 _ 로 변경합니다.
# 확장자와 구분이 쉽게 되도록 _ 로 변경합니다. 

last_day= df.iloc[0]['확진일']
last_day=last_day.replace('.',"_")
last_day
```

```python
# 파일명을 만들어 줍니다.
# file_name
file_name= f"seoul_covid19_{last_day}.csv" #WHY MUST PUT f in front?
file_name
```

```python
# csv 파일로 저장합니다.
df.to_csv(file_name,index=False, encoding= "utf-8")
```

```python
# 제대로 저장되었는지 확인합니다.
pd.read_csv(file_name)
```
